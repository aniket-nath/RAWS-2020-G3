```python
def plot_spec(file):
    # Code to stitch up the spectrum over diffrent frequency bands

    spect = pd.read_csv(file) # reading the csv file into panda dataframe
    stitch = [] # empty array for storing the stitched data
    for rows in spect.index: # looping through each frequency band 
        stitch.append(spect.loc[rows][6:]) # storing the intensity values
    l = len(stitch)
    l1 = len(stitch[0])
    final_spec = [] # empty array to store the final spectra in a single row
    for rows in stitch:
        # looping over each spectra band and appending the corresponding intensity
        for elements in rows:
            # looping and adding each intensity to the final array
            final_spec.append(elements)
    plt.plot(final_spec)
    
            

    
```
